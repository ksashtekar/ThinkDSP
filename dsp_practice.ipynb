```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
T1 = np.arange(0.0, 2 * 3.1416, 0.1)
V1 = np.array([math.sin(e) for e in T1])
V2 = np.array([math.sin(50 * e) for e in T1])
V3 = V1 * V2
plt.plot(T1, V3)
```
